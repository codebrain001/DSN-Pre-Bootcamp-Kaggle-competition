```python
cols_2_remove = ['Name','Ticket','Cabin','PassengerId']
n_data = data.drop(cols_2_remove, axis = 1)
n_data.head(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# The codes in this cell split the data into test data and train data: 70% train & 30% test
x_train, x_test, y_train, y_test = train_test_split(scaledD, y_target, test_size = 0.3, random_state = 400)
```

```python
print('Training data is ',x_train.shape)
print('Training target is',y_train.shape)
print('Test data is', x_test.shape)
print('Test target is',y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression(tol=0.0001, C=1000, fit_intercept=True, intercept_scaling=1, random_state=50, max_iter=1000, multi_class='ovr')
```

```python
help(LogisticRegression)
```

```python
logreg.fit(x_train,y_train)
```

```python
y_pred = logreg.predict(x_test)
y_pred
```

```python
from sklearn import metrics
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
#creating heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="rainbow",fmt='g')
ax.xaxis.set_label_position('top')
plt.tight_layout()
plt.title('confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show
```

```python
print("Accuracy: ", metrics.accuracy_score(y_test,y_pred))
print("Precision: ",metrics.precision_score(y_test,y_pred))
print("Recall: ", metrics.recall_score(y_test,y_pred))
```
